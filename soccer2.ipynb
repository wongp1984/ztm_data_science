```python
df1  = data_frame[["Name", "Wage", "Value"]]

```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    
    x = re.search(r'.(\d*\.*\d*[MKB]*)', x).group(1)
    
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
df1['C_Wage'] = df1[['Wage']].apply(lambda x: value_to_float(x['Wage']), axis=1)
df1['C_Value'] = df1[['Value']].apply(lambda x: value_to_float(x['Value']), axis=1)
# str1 = 'AA65.44'

# df1.Wage.str.extract(r'.(\d*\.*\d*[MKB]*)', expand=False)

# df1[['Wage']].apply(lambda x: value_to_float(x['Wage'].str.extract(r'.(\d*\.*\d*[MKB]*)', expand=False)), axis=1)

# print(type(df1.Wage.str.extract(r'.(\d*\.*\d*[MKB]*)', expand=False)))

# print(type(df1['Wage']))

# str1 = '$666.888M'
# print(re.search(r'.(\d*\.*\d*[MKB]*)', str1).group(1))
```
